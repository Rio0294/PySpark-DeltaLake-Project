```python
from pyspark.sql import SparkSession

spark = (
        SparkSession
        .builder
        .appName("Read csv")
        .master("local[*]")
        .getOrCreate()
        
)
spark
```

```python
#Read csv in dataframe

df = spark.read.format("csv").option("header", True).option("inferSchema", True).load(r"C:\CODING_RIYA_WORK\emp.csv")
```

```python
df.printSchema()
```

```python
df.show()
```

```python
_schema = "employee_id int, department_id int, name string, age int, gender string, salary double, hire_date date"

df_schema = spark.read.format("csv").option("header", True).schema(_schema).load(r"C:\CODING_RIYA_WORK\emp.csv")
```
