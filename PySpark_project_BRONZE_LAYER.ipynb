```python
#Initiate Spark Session
from pyspark.sql import SparkSession

import os
os.environ["HADOOP_HOME"] = "C:\\hadoop"
os.environ["hadoop.home.dir"] = "C:\\hadoop"

import sys
sys.path.append("C:\\hadoop\\bin")

spark = SparkSession.builder.appName("PyProject").master("local[*]").getOrCreate()

spark
```

```python
#Read csv file


file_path = r"C:\CODING RIYA WORK\retail_store_sales.csv"

sales = spark.read.csv(file_path)
```

```python
#Call an action to show the csv file
sales.show()
```

```python
#Reading csv file into Spark Dataframe
sales_df = spark.read.csv(file_path, header=True)
```

```python
sales_df.printSchema()
```

```python
#Call an action to show the dataframe 
sales_df.show()
```

```python
#Add ingestion timestamp
from pyspark.sql.functions import current_timestamp
retail_final = sales_df.withColumn("datetime_now", current_timestamp())

#Call action to show the dataframe
retail_final.show(truncate=False)
```

```python
# #Write as Parquet

output_path = r"C:\CODING_RIYA_WORK\output"
retail_final.write.mode("overwrite").option("header", True).parquet(output_path)
```
